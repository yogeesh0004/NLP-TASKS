```python
from urllib import request
```

```python
url = "https://www.gutenberg.org/cache/epub/69938/pg69938.txt"
```

```python
response = request.urlopen(url)
```

```python
raw=response.read().decode('utf8')
```

```python
import nltk
from nltk.tokenize import word_tokenize
tokens=word_tokenize(raw)
```

```python
print(tokens[:23])
```

```python
#partsofspeechtagging
tagged=nltk.pos_tag(tokens)
```

```python
print(tagged)
```

```python
import nltk
from nltk.stem import RegexpStemmer
reg = RegexpStemmer('ing$')
reg.stem('walking')
```

```python
import nltk
from nltk.stem import SnowballStemmer
snow=SnowballStemmer('french')
snow.stem('manges')
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
ps=PorterStemmer()
words=["program","programs","programmer","programmers"]
for w in words:
    print(w,":",ps.stem(w))
```

```python
porter = PorterStemmer()
text="""This eBook is for the use of anyone anywhere in the United States and
most other parts of the world at no cost and with almost no restrictions
whatsoever. You may copy it, give it away or re-use it under the terms
of the Project Gutenberg License included with this eBook or online at
www.gutenberg.org. If you are not located in the United States, you
will have to check the laws of the country where you are located before
using this eBook.
"""
stemmed=[porter.stem(token)for token in text.split(" ")]
print(stemmed)
```
